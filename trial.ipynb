```python
# --- configuration -------------------------------------------------
SIZE = 3                          # 3×3 board; change to 4, 5 … for larger grids
EMPTY = " "                       # symbol for an unused square


# create an empty SIZE×SIZE matrix
board = [[EMPTY for _ in range(SIZE)] for _ in range(SIZE)]

def print_board(b):
    """
    Print the board with row (R) and column (C) headers:

        |R\C| 0 | 1 | 2 |
        ------------------
        | 0 |   | X |   |
        ------------------
        | 1 |   |   |   |
        ------------------
        | 2 |   |   |   |
        ------------------
    """
    # column header
    header = " |R\\C| " + " | ".join(str(c) for c in range(SIZE)) + " |"
    print(header)
    print("-" * len(header))

    # each row with its index
    for r in range(SIZE):
        row_cells = " | ".join(b[r])
        print(f"| {r} | {row_cells} |")
        print("-" * len(header))

def place_mark(b, row, col, mark):
    """
    Put `mark` at (row, col) if the indices are in range
    and the square is empty. Modify as needed for error handling.
    """
    if 0 <= row < SIZE and 0 <= col < SIZE:
        if b[row][col] == EMPTY:
            b[row][col] = mark
        else:
            raise ValueError("Square already occupied")
    else:
        raise IndexError("Row or column out of bounds")

# ----- demonstration ------------------------------------------------
print_board(board)            # empty grid
place_mark(board, 0, 1, "X")  # put X at row 0, column 1
print_board(board)
```

```python
class TicTacToe:
    """
    simple tic tac toe 3 x 3 board game

    how use:
        game = TicTacToe()
        game.play()  # start interactive game loop with prompt
    """

    BOARD_HEIGHT = 6
    BOARD_WIDTH=7
    EMPTY_SYMBOL = " "

    def __init__(self):
        # prepare empty board with blanks
        self.play_area = []
        for row_counter in range(self.BOARD_HEIGHT + 1):
            temp_row = []
            for col_counter in range(self.BOARD_WIDTH + 1):
                temp_row.append(self.EMPTY_SYMBOL)
            self.play_area.append(temp_row)

    def display_board(self):
        # print column headers
        header_line = "|R\\C| "
        
        # print rows with indices and board cells
        for row_number in range(self.BOARD_HEIGHT,0,-1):
            line_print = "| " + str(row_number) + " | "
            for col_number in range(self.BOARD_WIDTH,0,-1):
                # print(row_number,col_number)
                line_print += self.play_area[row_number][col_number] + " | "
            print(line_print)
            separator = "-" * len(line_print)
            print(separator)
        
        for col_number in range(self.BOARD_WIDTH):
            header_line += str(chr(97+col_number)) + " | "
        print(header_line)
        return self.play_area

    def input_valid(self, row, col):
        # check if coordinates are on board and spot free
        if row < 0 or row >= self.BOARD_SIDE:
            return False
        if col < 0 or col >= self.BOARD_SIDE:
            return False
        if self.play_area[row][col] != self.EMPTY_SYMBOL:
            return False
        return True
    
    def valid_places(self):
        pass

    def is_board_full(self):
        # returns True if no spot empty found
        for r in range(self.BOARD_SIDE):
            for c in range(self.BOARD_SIDE):
                if self.play_area[r][c] == self.EMPTY_SYMBOL:
                    return False
        return True

    def has_won(self, mark):
        # check for mark winning any row
        for r in range(self.BOARD_SIDE):
            is_winner = True
            for c in range(self.BOARD_SIDE):
                if self.play_area[r][c] != mark:
                    is_winner = False
                    break
            if is_winner:
                return True

        # check for mark winning any column
        for c in range(self.BOARD_SIDE):
            is_winner = True
            for r in range(self.BOARD_SIDE):
                if self.play_area[r][c] != mark:
                    is_winner = False
                    break
            if is_winner:
                return True

        # check major diagonal
        is_winner = True
        for i in range(self.BOARD_SIDE):
            if self.play_area[i][i] != mark:
                is_winner = False
                break
        if is_winner:
            return True

        # check anti-diagonal
        is_winner = True
        for i in range(self.BOARD_SIDE):
            if self.play_area[i][self.BOARD_SIDE - 1 - i] != mark:
                is_winner = False
                break
        if is_winner:
            return True

        return False

    def game_finished(self, mark):
        # returns True if either player wins or board is filled (draw)
        if self.has_won(mark):
            return True
        if self.is_board_full():
            return True
        return False

    def play(self):
        # run interactive game loop
        self.clear_board()
        print("New Game: X goes first.\n")
        self.display_board()

        current_mark = "X"
        while True:
            print(f"\n{current_mark}'s turn.")
            user_move = input(
                f"Where do you want your {current_mark} placed?\n"
                "Please enter row number and column number separated by a comma.\n"
            )

            try:
                row_str, col_str = user_move.split(",")
                row_input = int(row_str.strip())
                col_input = int(col_str.strip())
                print(f"You have entered row #{row_input}\n\tand column #{col_input}\n")
                print("Thank you for your selection.")
            except Exception:
                print("Invalid input format. Try again.")
                continue

            if not self.input_valid(row_input, col_input):
                if 0 <= row_input < self.BOARD_SIDE and 0 <= col_input < self.BOARD_SIDE:
                    print("That cell is already taken.")
                else:
                    print(
                        "Invalid entry: try again.\n"
                        "Row & column numbers must be either 0, 1, or 2."
                    )
                continue

            self.play_area[row_input][col_input] = current_mark
            self.display_board()

            if self.has_won(current_mark):
                print(f"\n{current_mark} IS THE WINNER!!!")
                break
            if self.is_board_full():
                print("\nDRAW! NOBODY WINS!")
                break

            # flip turn
            current_mark = "O" if current_mark == "X" else "X"

        reply = input("\n Another game? Enter Y or y for yes.\n")
        if reply.lower() == "y":
            self.play()
        else:
            print("Thank you for playing!")



if __name__ == "__main__":
    game_instance = TicTacToe()
    game_instance.play()
```
